주택구매력지수란 중위 소득을 가진 가구가 금융기관의 대출을 받아 중간가격 정도의 주택을 구입한다고 가정할 때, 원리금 상환에 필요한 금액을 부담할 수 있는 능력을 의미합니다.

주택구매력지수(HAI) = (중위가구소득 ÷ 대출 상환 가능 소득) ×100

- KB주택구입잠재력지수
- HOI = (구입 가능 재고량 ÷ 총 재고량) ×100

주택구입잠재력지수(KB-HOI) 란 중위 소득의 가구가 금융기관의 대출을 받아 주택을 구입한다고 가정할 경우, 지역 내 주택 중 경제능력(소득, 자산 등)의 제약 속에서 구입 가능한 주택 재고량을 알 수 있는 지표입니다.
즉, KB-HOI 지수가 50이면 해당 지역 내 주택 재고 수의 가격별 하위 50% 범위 내 주택을 구입할 수 있음을 나타내며, KB-HOI 지수가 상승하면 주택 구입 능력(잠재력)이 증가한다는 것을 뜻합니다.
HOI = (구입 가능 재고량 ÷ 총 재고량) ×100

- KB아파트주택담보대출 PIR
- KB-PIR = 주택 가격 / (연간)가구 소득

KB국민은행 부동산담보대출(아파트) 대출자를 대상으로 한 주택 가격을 가구 소득으로 나눈 통계입니다. KB-PIR = 주택 가격 / (연간)가구 소득
가구 소득: 분기단위 해당 지역 내 KB국민은행 부동산담보대출(아파트) 대출자의 연소득 중위값
주택 가격: 분기단위 해당 지역 내 KB국민은행 부동산담보대출(아파트) 실행 시 조사된 담보평가 가격의 중위가격

```python
api = Kbland()
params = {
    "기간": "1"
}
df = api.get_apartment_mortgage_loan_pir(**params)
df.tail()
```

```python
# 지역명이 "경기"인 데이터 필터링
경기_data = df[df["지역명"] == "경기"]

print(f"경기 데이터 총 개수: {len(경기_data)}")
print("\n경기 데이터 컬럼:")
print(경기_data.columns.tolist())
print("\n경기 데이터 샘플:")
경기_data.head()
```

```python
# 경기 데이터를 주택분위와 소득분위별로 정렬 (내림차순: 5,4,3,2,1)
경기_sorted = 경기_data.sort_values(['주택분위', '소득분위'], ascending=[False, False])

print("경기 - 주택분위별, 소득분위별 정렬 (5,4,3,2,1):")
print("="*50)
경기_sorted
```

```python
# 주택분위별로 나누어 보기
print("경기 - 주택분위별 상세 분석:")
print("="*60)

for 주택분위 in [5, 4, 3, 2, 1]:
    해당분위_data = 경기_sorted[경기_sorted['주택분위'] == 주택분위]
    if len(해당분위_data) > 0:
        print(f"\n📊 주택분위 {주택분위}분위:")
        print(f"   데이터 개수: {len(해당분위_data)}개")
        print(f"   소득분위별 데이터:")
        
        # 소득분위별로 PIR 값 출력
        for 소득분위 in [5, 4, 3, 2, 1]:
            해당소득_data = 해당분위_data[해당분위_data['소득분위'] == 소득분위]
            if len(해당소득_data) > 0:
                pir_값 = 해당소득_data['PIR'].iloc[0]
                print(f"      소득분위 {소득분위}: PIR = {pir_값:.2f}")
        print("-" * 40)
```

```python
# 최신 데이터만 보기 (가장 최근 날짜)
최신날짜 = 경기_data['날짜'].max()
최신_경기_data = 경기_data[경기_data['날짜'] == 최신날짜]

print(f"경기 - 최신 데이터 ({최신날짜}):")
print("="*50)

# 최신 데이터 피벗 테이블
최신_pir_matrix = 최신_경기_data.pivot_table(
    index='주택분위', 
    columns='소득분위', 
    values='PIR', 
    aggfunc='first'
)

# 내림차순 정렬 (5,4,3,2,1)
최신_pir_matrix = 최신_pir_matrix.sort_index(ascending=False)
최신_pir_matrix = 최신_pir_matrix.reindex(sorted(최신_pir_matrix.columns, reverse=True), axis=1)

print("📅 최신 PIR 매트릭스:")
print("   주택분위(행) x 소득분위(열)")
최신_pir_matrix
```

```python
# 히트맵으로 시각화
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 폰트 설정
plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False

# 히트맵 생성
plt.figure(figsize=(10, 6))
sns.heatmap(최신_pir_matrix, 
            annot=True, 
            fmt='.2f', 
            cmap='YlOrRd',
            cbar_kws={'label': 'PIR 값'})

plt.title(f'경기 PIR 히트맵 ({최신날짜})', fontsize=16, fontweight='bold')
plt.xlabel('소득분위', fontsize=12)
plt.ylabel('주택분위', fontsize=12)
plt.tight_layout()
plt.show()

print(f"\n💡 해석:")
print(f"- PIR 값이 높을수록 주택 구입이 어려움")
print(f"- 주택분위 5 (고가주택) + 소득분위 1 (저소득)이 가장 높은 PIR")
print(f"- 주택분위 1 (저가주택) + 소득분위 5 (고소득)이 가장 낮은 PIR")
```
